## Диаграммы размаха для проверки наличия выбросов среди переменных

Диаграмма "ящик с усами" (диаграмма размаха, коробчатая диаграмма, блочная диаграмма) с ограничителями выбросов — это график, использующийся в описательной статистике, компактно изображающий одномерное распределение вероятностей.

Такой вид диаграммы в удобной форме показывает медиану (или, если нужно, среднее), нижний и верхний квантили, минимальное и максимальное значение выборки и выбросы. 

```python
fig, axes = plt.subplots(4, 4, figsize=(15, 5*4))
fig.subplots_adjust(hspace=1) # Регулирование расстояния по вертикали

a = 0
for i in range(4):
    for j in range(4):
        ax = axes.flatten()[a]
        sns.boxplot(x=data[data.columns[a]], ax=ax, color=colors[a])
        #Установить метки оси и заголовок
        ax.set_xlabel(data.columns[a])
        ax.set_title(data.columns[a])
        a += 1
        
# добавляем общий заголовок
plt.suptitle('Диаграммы размаха («ящик с усами») для обнаружения выбросов \n',  fontsize=19)        
#Регулировка макета, чтобы предотвратить наложение 
plt.tight_layout()
plt.show()
```

## Графики для дискретных величин

Проведем подсчет домов по таким признакам, как уровень шума, месяц продажи, качество конструкции

```python
for feature in discrete_feature:
    plt.figure(figsize=[6, 4])
    plot = sns.countplot(data=data, x=feature)
    total = len(data[feature])
    for p in plot.patches:
        percentage = '{:.1f}%'.format(100 * p.get_height()/total)
        x = p.get_x() + p.get_width() / 2 - 0.05
        y = p.get_y() + p.get_height()
        plot.annotate(percentage, (x, y), size = 10)
        plot.set(ylabel='количество домов')
    
    plt.show() 
```

**Вывод**:

Q-Q график - это график, на котором квантили из двух распределений расположены относительно друг друга. Чем ближе точки на графике к диагональной прямой, тем ближе распределение исследуемой переменной к нормальному закону.
График QQ используется для оценки того, потенциально ли набор данных получен из некоторого теоретического распределения. В большинстве случаев этот тип графика используется для определения того, соответствует ли набор данных нормальному распределению.

Если данные распределены нормально, точки на графике QQ будут лежать на прямой диагональной линии.

И наоборот, чем больше точки на графике отклоняются от прямой диагональной линии, тем менее вероятно, что набор данных следует нормальному распределению.

На диагонали матриц будет стоять дисперсия оценок коэффициентов, а числа вне главной диагонали равна их ковариациям. С помощью этой матрицы мы можем сделать вывод о линейной зависимости коэффициентов регрессии друг от друга.


**Результаты модели**:
  
1. R-squared - сколько процентов дисперсии зависимой переменной объяснено вариацией признаков.
2. Adj. R-squared - та же метрика, но со штрафом за большое количество зависимых переменных. При добавлении новых признаков значение R-Squared будет таким же или расти, даже если они совсем не влияют на результат. Данная метрика штрафует за использование каждого дополнительного признака в модели.
3. F-statistic - критерий Фишера - статистика, значение которой необходимо для проверки гипотезы о равенстве дисперсий. В данном случае сравнивается модель с нулевыми параметрами (кроме интерсепта) с текущей.
4. Prob. (F-statistic) - вероятность того, что коэффициенты при всех переменных равны нулю (кроме интерсепта).
5. Log-likelihood - насколько хорошо данные описываются моделью. Важно понимать, что по абсолютному значению правдоподобия нельзя делать никаких вероятностных суждений. Правдоподобие позволяет сравнить несколько вероятностных распределений с разными параметрами и оценить в контексте какого из них наблюдаемые события наиболее вероятны.
6. AIC BIC - используется для сравнения моделей. Как таковое значение не несёт в себе никакой информации, его можно использовать как критерий для отбора модели, который определяет на сколько сильно модель переобучена. Чем меньше значение - тем лучше модель. BIC отличается от AIC тем, что он больше штрафует за дополнительные (и не влияющие ни на что) параметры модели.
7. Intercept - свободный член. Все переменные обнуляются и подбирается константа, наилучшим образом описывающая данные.
8. std err - дисперсия коэффициента по точкам данных.
9. t - статистика t-Стьюдента, чем больше — тем лучше измерен коэффициент (чем меньше стандартное отклонение, тем лучше коэффициент описывает зависимость). Обычно t-критерий позволяет нам оценить важность различных предикторов, предполагая, что остатки модели нормально распределены около нуля. Если остатки не ведут себя таким образом, то это говорит о наличии некоторой нелинейности между переменными и о том, что их t-тесты не следует использовать для оценки важности отдельных предикторов.
10. P>|t| - если < 0.05, то гипотеза о значимости коэффициента принимается. Например, переменная Literacy не влияет на зависимую переменную Loterry с вероятностью 37.8% и наши результаты получены случайно.
11. Доверительные интервалы коэффициента. Меньший доверительный интервал предполагает, что мы уверены в значении оцениваемого коэффициента или постоянного члена. И наоборот. В данном случае показываются значения, при которых гипотеза о значимости коэффициента не будет отвергнута.
    
    
**Анализ остатков**
    
1. Omnibus описывает нормальность распределения наших остатков (остатки необъяснённой дисперсии), используя в качестве измерений перекос (skew) (коэффициент симметрии распределения) и эксцесс (kurtosis) (мера остроты случайной величины в горбу). 0 означает полную нормальность.
2. Prob (Omnibus) - это статистический тест, измеряющий вероятность нормального распределения остатков. Значение 1 означает совершенно нормальное распределение.
3. Skew (Перекос) - это мера симметрии распределения остатков, где 0 означает идеальную симметрию
4. Kurtosis (Эксцесс) - измеряет остроту распределения остатков (в горбу) или его концентрацию около 0 на нормальной кривой. Более высокий эксцесс означает меньшее количество выбросов.
5. Durbin-Watson - это критерий для проверки наличия автокорреляции. При отсутствии автокорреляции значение критерия находится между 1 и 2.
6. Jarque-Bera (JB) и Prob (JB) - альтернативные методы измерения того же значения, что и Omnibus и Prob (Omnibus), с использованием асимметрии (перекос) и эксцесса. Нулевая гипотеза – распределение является нормальным, асимметрия равна нулю, а эксцесс равен трем. При небольших выборках тест Jarque-Bera склонен отклонять нулевую гипотезу когда она верна.
7. Cond. No (число обусловленности) — это мера чувствительности нашей модели по отношению к входящим данным. То есть при малейшем изменении данных коэффициенты (соответственно и отклик модели) сильно меняется. Мультиколлинеарность следует из большого числа обусловленности. 
 

# Удаление выбросов 

Для улучшения качества модели избавимся от выбросов (аномальных значений в датасете). Для этого рассчитаем значения стандартных остатков и удалим записи, в которых они >2 по модулю

```python
coefs = model.params
coefs.to_list() # коэффициенты регрессии
x = x.drop(['const'], axis=1)
teor_znach = []
cols = list(x.columns)
for i in range(len(y)):
    a = coefs[0]
    for j in range(len(cols)):
        a += x[cols[j]][i] * coefs[j+1]
    teor_znach.append(a)
```

```python
ostatki = []
for i in range(len(teor_znach)):
    ostatki.append(y[i] - teor_znach[i])
```

```python
ost_disp = sum(list(map(lambda x: x**2, ostatki)))
print(ost_disp)
```

```python
n = len(y)
standart_ostatki = list(map(lambda x: x/(ost_disp/(n - 1))**0.5, ostatki))
print(f"Стандартные остатки: {standart_ostatki[:10]}...")
```

```python
# Удаление записей с выбросами
clean_x = x
clean_y = y
for i in range(len(y)):
    if abs(standart_ostatki[i]) > 2:
        clean_x = clean_x.drop(index=i)
        clean_y = clean_y.drop(index=i)
```

## Тест Глейзера

Тест Глейзера — статистический тест, позволяющий оценить наличие (отсутствие) гетероскедастичности случайных ошибок регрессионной (эконометрической) модели.

Позволяет выявить зависимость модуля остатков от функциональных форм каждого регрессора. Выдвигаются гипотезы:

* H0: остатки гомоскедастичны (β = 0)
* H1: остатки гетероскедастичны

**Алгоритм:**
* Строим исходную регрессию по всем имеющимся данным и сохраняем остатки

## Тест Колмогорова-Смирнова

Проверим выполнение предосылки с помощью теста Колмогорова-Смирнова.

Тест Колмогорова-Смирнова используется для проверки того, подчиняется ли выборка определенному распределению, в нашем случае - нормальному.

Чтобы выполнить тест Колмогорова-Смирнова в Python, воспользуемся методом **scipy.stats.kstest() для теста с одной выборкой**.

```python
stats.kstest(clean_data["ε"], 'norm')
```

Статистика теста равна 0.566, а соответствующее p-значение равно 0.0. Поскольку p-значение меньше 0.05, мы отвергаем нулевую гипотезу о нормальном распределении выборки. Таким образом, данные выборки, т.е. остатки, не имеют нормального распределения.
